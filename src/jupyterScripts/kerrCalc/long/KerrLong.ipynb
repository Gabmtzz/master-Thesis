```fortran
program KerrLong
implicit none
integer,parameter::Ntot=300
real,parameter::PI=3.141592653589793
real,parameter::theta=60.0
real::RadTheta
complex,dimension(Ntot)::exx,ezz,exz
real,dimension(Ntot)::omega

print *, '=========================================================================='
print *, 'programa para calcular espectro de efecto Kerr sin aplicar campo magnetico'
print *, '=========================================================================='

print *, 'leyendo archivos .dat'
! obtener exx.dat
call CrearArr('epsxx.dat',omega,exx)
! obtener ezz.dat
call CrearArr('epszz.dat',omega,ezz)
! obtener exz.dat
call CrearArr('epsxz.dat',omega,exz)

RadTheta=(theta*PI)/180
print *,'calcular espectro Kerr'
call cKerrLong(exx,exz,ezz,RadTheta,omega)

end program KerrLong

!=============================================================================================================================

subroutine cKerrLong (exx,exz,ezz,RadTheta,omega)
implicit none 
integer,parameter::Ntot=300
complex,dimension(Ntot)::Darr,exx,ezz,exz,KerrComp
real,dimension(Ntot)::thetaK,etaK,omega
integer::i
real::RadTheta

call D(exx,exz,ezz,RadTheta,Darr)
!calcula angulo kerr complejo
KerrComp =(-1)*(2*exz*cos(radTheta)*sin(radTheta)*sqrt(exx)/Darr)
!obtiene angulo y elipticidad kerr
call ReImag(Kerrcomp,thetaK,etaK)

!escribir archivo
print *,'Guardando informacion'
open(unit=13,file='KerrLong_PtS2.dat')

do i=1,Ntot
    write(13,*)omega(i),thetaK(i),etaK(i)
end do
close(13)

end subroutine
!+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
subroutine D (exx,exz,ezz,RadTheta,Darr)
implicit none
integer,parameter::Ntot=300
complex,dimension(Ntot)::Darr,exx,ezz,exz
real::RadTheta

Darr=(sqrt(exx*(ezz-sin(RadTheta)**2))+sqrt(ezz*(exx-sin(RadTheta)**2)))*(sqrt(exx-sin(RadTheta)**2)+cos(RadTheta))&
    *(sqrt(exx*ezz)*cos(RadTheta)-sqrt(ezz-sin(RadTheta)**2))

end subroutine
!==========================================================================================================
subroutine CrearArr (Nombre,omega,compArr)
implicit none
integer,parameter::Ntot=300
real,dimension(Ntot)::omega
character(len=9)::Nombre
complex,dimension(Ntot)::compArr
real,dimension(Ntot)::reArr,ImArr

integer::i
open(unit=13, file=Nombre)
do i=1,Ntot
   read(13,*)omega(i),reArr(i),ImArr(i)
end do
close(13)
call complx(reArr,ImArr,compArr)
end subroutine 
!==========================================================================================================
! crear un arreglo  complejo

subroutine complx (reArr,ImArr,compArr)
implicit none
integer,parameter::Ntot=300
real,dimension(Ntot)::reArr,ImArr
complex,dimension(Ntot)::compArr

compArr=CMPLX(reArr, ImArr)
end subroutine

!+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
! separa un arreglo  complejo

subroutine ReImag (compArr,reArr,ImArr)
implicit none
integer,parameter::Ntot=300
real,dimension(Ntot)::reArr,ImArr
complex,dimension(Ntot)::compArr

reArr=REAL(compArr)
ImArr=AIMAG(compArr)
end subroutine
```
